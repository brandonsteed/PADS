```python
df['long_commute'] = np.where(df['MeanCommute']>=30, 1, 0)
```

```python
df['short_commute'] = np.where(df['MeanCommute']<=5, 1, 0)
```

```python
df['high_public_work'] = np.where(df['PublicWork']>=20, 1, 0)
```

```python
df['high_private_work'] = np.where(df['PrivateWork']>=80, 1, 0)
```

```python
df['low_public_work'] = np.where(df['PublicWork']<=10, 1, 0)
```

```python
df['low_private_work'] = np.where(df['PrivateWork']<=70, 1, 0)
```

```python
df['employed_outside_home'] = df['PublicWork'] + df['PrivateWork']
```

```python
features = pd.get_dummies(df.TotalPop > 100000, prefix='PopOver100000')
```

```python
features = pd.concat([df, features], axis=1)
```

```python
features2 = pd.get_dummies(df.TotalPop < 20000, prefix='PopUnder20000')
```
